```python
import pymongo
from pymongo import MongoClient

client = MongoClient()

Querydb = client['Query'] #Define Database

quote_collection = Querydb['Quotes'] #Define Quote Collection
word_collection =  Querydb['Words'] #Define Words Collection
```

```python
import pandas as pd

words = pd.DataFrame(pd.read_csv('C:/Users/Prateek/Desktop/Words.csv',skiprows=[1]))

for word in words.iterrows():
    word_collection.insert_one({'word_id':word[1][0],'search_word':word[1][1]})
```

```python
for row in word_collection.find():
    print(row)
```

```python
quotes = pd.DataFrame(pd.read_csv('C:/Users/Prateek/Desktop/Quotes.csv',skiprows=[0]))

for row in quotes.iterrows():
    quote_collection.insert_one({'quote_id':row[1][0],'author':row[1][1], 'quote':row[1][2]})
```

```python
for quote in quote_collection.find():
    print(quote)
```

```python
search_word = [word['search_word'] for word in word_collection.find()]
```

```python
single_sentence = ' '.join(val['quote'] for val in quote_collection.find()).lower()
```

```python
word_count = {word:single_sentence.count(word) for word in search_word}
```
